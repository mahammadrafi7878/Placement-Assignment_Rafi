```python
import statsmodels.api as sm
import pylab as py              #using Q-Q plot
sm.qqplot(df,line='45')      #They are not in same line ,so not norally distributed
py.show()
```

```python
from scipy import stats               #using statistical concepts    k-square
alpha=0.05
for i in df.columns:
    print([i])
    a,b=stats.normaltest(df[[i]])
    print(a,b)
    if b<alpha:
        print("reject")
    else:
        print("not reject")
```
